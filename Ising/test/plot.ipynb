```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mag_list = [-0.75, -0.5, -0.25, 0.0, 0.25, 0.5, 0.75]

# Load test_latent
with open('test_latent.pkl', 'rb') as f:
    test_latent = pickle.load(f)

# Load predict_ours
with open('predict_ours.pkl', 'rb') as f:
    predict_ours = pickle.load(f)

# Load predict_naive
with open('predict_naive.pkl', 'rb') as f:
    predict_naive = pickle.load(f)
```

```python
mean_dt = 6.2021
t_mean = np.arange(0, 500) * mean_dt
```

```python
# ========== Enhanced Academic Visualization for Publication ==============
j = 0  # Using first magnetization condition for detailed analysis
true_tra = test_latent[f'box_L_16_mag_' + str(mag_list[j])]
pred_tra_ours = predict_ours['box_L_16' + '_' + 'mag_' + str(mag_list[j])]
pred_tra_naive = predict_naive['box_L_16' + '_' + 'mag_' + str(mag_list[j])]

true_mean = np.mean(true_tra, axis=0)
pred_mean_ours = np.mean(pred_tra_ours, axis=0)
pred_mean_naive = np.mean(pred_tra_naive, axis=0)

# Calculate statistical measures
true_std = np.std(true_tra, axis=0)
pred_std_ours = np.std(pred_tra_ours, axis=0)
pred_std_naive = np.std(pred_tra_naive, axis=0)
# Define professional academic color scheme for publications
colors = {
    'true': 'tab:blue',      # Near black for ground truth (authoritative)
    'ours': 'tab:orange',      # Strong red for our method (attention-grabbing)
    'naive': 'tab:grey',     # Professional blue for baseline (standard)
}
```

```python
plt.rcParams.update({
    'font.size': 36,
    'font.family': 'sans-serif', 
    'text.usetex': False,  # Set to True if LaTeX is available
    'axes.linewidth': 1.2,
    'axes.spines.right': False,
    'axes.spines.top': False,
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.minor.size': 2.5,
    'ytick.minor.size': 2.5,
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'legend.frameon': False,
    'legend.fontsize': 30,
})
# Create figure with academic styling
fig = plt.figure(figsize=(16, 6), facecolor='white', dpi=300)

# Individual trajectory comparisons - Dimension 1
axes1 = fig.add_subplot(1, 2, 1)
axes1.plot(t_mean, pred_mean_ours[:, 0], color=colors['ours'], label='Our method', 
           linewidth=4, linestyle='-', zorder=3)
axes1.plot(t_mean, pred_mean_naive[:, 0], color=colors['naive'], label='Baseline', 
           linewidth=4, linestyle='-', zorder=2)
axes1.plot(t_mean, true_mean[:, 0], color=colors['true'], label='Ground Truth', 
           linewidth=4, linestyle='-', zorder=1)
axes1.legend(loc='lower right')
axes1.set_ylabel('Mean')
axes1.set_xlabel('Time')

axes2 = fig.add_subplot(1, 2, 2)
axes2.plot(t_mean, pred_std_ours[:, 0], color=colors['ours'], label='Our method', 
           linewidth=4, linestyle='-', zorder=3)
axes2.plot(t_mean, pred_std_naive[:, 0], color=colors['naive'], label='Baseline', 
           linewidth=4, linestyle='-', zorder=2)
axes2.plot(t_mean, true_std[:, 0], color=colors['true'], label='Ground Truth', 
           linewidth=4, linestyle='-', zorder=1)
axes2.set_ylabel('Std')
axes2.set_xlabel('Time')

# Adjust layout for academic presentation
plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.suptitle('(a) Magnetization', y=0.98, fontsize=32)

plt.savefig('ising_traj_1.png', dpi=300, bbox_inches='tight', 
            facecolor='white', edgecolor='none')
plt.savefig('ising_traj_1.pdf', bbox_inches='tight', 
            facecolor='white', edgecolor='none')
plt.show()
```

```python
# ========== Enhanced Academic Visualization for Ablation Study ==============

# Set up professional academic figure style
plt.rcParams.update({
    'font.size': 36,
    'font.family': 'sans-serif',
    'text.usetex': False,  # Set to True if LaTeX is available
    'axes.linewidth': 1.2,
    'axes.spines.right': False,
    'axes.spines.top': False,
    'xtick.major.size': 6,
    'ytick.major.size': 6,
    'xtick.minor.size': 3,
    'ytick.minor.size': 3,
    'legend.fontsize': 32,
    'legend.frameon': False,
})
# Define academic color scheme
colors = {
    'ours': '#D62728',      # Professional blue
    'naive': '#1F77B4',     # Contrasting orange
}

# Create figure with proper academic dimensions
fig = plt.figure(figsize=(16, 6), dpi=300, facecolor='white')
axes1 = fig.add_subplot(1, 2, 1)
plt.xscale('log')

# Enhanced plotting with error bars and better markers
line1 = axes1.plot(x_labels_CW**2, mse_ours_CW, 
                   marker='o', color=colors['ours'], 
                   label='Our method', 
                  markersize=10, linewidth=4, 
                  markerfacecolor='white', markeredgewidth=2,
                  markeredgecolor=colors['ours'])

line2 = axes1.plot(x_labels_CW**2, mse_naive_CW, 
                  marker='s', color=colors['naive'], 
                  label='Baseline', 
                  markersize=10, linewidth=4,
                  markerfacecolor='white', markeredgewidth=2,
                  markeredgecolor=colors['naive'])

# Enhanced axis labels with proper mathematical notation
axes1.set_xlabel('$n_s$')
axes1.set_ylabel('Test Error')
axes1.invert_xaxis()

# Set ticks and formatting
x_tick_values = x_labels_CW**2
x_tick_labels = [f'{int(x)}²' for x in x_labels_CW]
axes1.set_xticks(x_tick_values)
axes1.set_xticklabels(x_tick_labels)

axes1.tick_params(axis='y')
axes1.set_title('(a) Curie-Weiss Model', pad=15, fontsize=32)
legend = axes1.legend(loc='upper left')


axes2 = fig.add_subplot(1, 2, 2)

# Enhanced plotting with error bars and better markers
line1 = axes2.plot(x_labels_ising**2, mse_ours_ising, 
                   marker='o', color=colors['ours'], 
                   label='Our method', 
                  markersize=10, linewidth=4, 
                  markerfacecolor='white', markeredgewidth=2,
                  markeredgecolor=colors['ours'])

line2 = axes2.plot(x_labels_ising**2, mse_naive_ising, 
                  marker='s', color=colors['naive'], 
                  label='Baseline', 
                  markersize=10, linewidth=4,
                  markerfacecolor='white', markeredgewidth=2,
                  markeredgecolor=colors['naive'])

# Enhanced axis labels with proper mathematical notation
axes2.set_xlabel('$n_s$')
plt.xscale('log')

# Set ticks and formatting
x_tick_values = x_labels_ising**2
x_tick_labels = [f'{int(x)}²' for x in x_labels_ising]
axes2.set_xticks(x_tick_values)
axes2.set_xticklabels(x_tick_labels)
axes2.tick_params(axis='y')
axes2.invert_xaxis()
axes2.set_title('(b) Ising Model', pad=15, fontsize=32)
legend = axes2.legend(loc='upper left')

# Tight layout for proper spacing
plt.tight_layout()
# plt.tight_layout(pad=15, w_pad=0.2, h_pad=0.2)
plt.savefig('CW_Ising.png', dpi=300, bbox_inches='tight', 
            facecolor='white', edgecolor='none')
plt.savefig('CW_Ising.pdf', bbox_inches='tight', 
            facecolor='white', edgecolor='none')

# Display the plot
plt.show()
```
